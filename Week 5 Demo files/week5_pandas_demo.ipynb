## Step 1 — Load the data and take a first look

`pd.read_csv()` loads a CSV file into a DataFrame. `df.head()` shows the first five rows. This is always your first move.

### What is a DataFrame?

A **DataFrame** is pandas' main way of storing tabular data. You can think of it like a spreadsheet or a CSV loaded into Python: it has rows, columns, and column names.

Pandas uses DataFrames because they make it easier to inspect, filter, summarize, clean, and analyze data with short commands. Instead of manually looping through every row in a CSV file, you can ask questions like:

```python
df.head()
df['rating'].mean()
df[df['rating'] < 3]
```

In this notebook, `df` is the name of our DataFrame. Once the CSV is loaded into `df`, every later step uses that DataFrame to explore the dataset.

````python
df = pd.read_csv('app_reviews_demo.csv')
df.head()
````

`df.tail()` shows the last five rows — useful for spotting truncated exports or checking whether the file ended cleanly.

````python
df.tail()
````

## Step 2 — Check the structure — `df.info()`

`df.info()` tells you:
- How many rows and columns you have
- The data type of each column (`int64`, `object`, `bool`, `float64`)
- How many **non-null** values are in each column

Look at the non-null count column carefully. Any column with fewer non-null values than total rows has missing data.

````python
df.info()
````

---
## Step 3 — Summarize the numbers — `df.describe()`

`df.describe()` gives you count, mean, min, max, and quartiles for every numeric column. Useful for spotting:
- Unexpected min or max values (outliers)
- Whether a column's range makes sense
- Whether the mean and median (50%) are far apart (skew)

````python
df.describe()
````

`describe()` skips non-numeric columns by default. Pass the `include='all'` parameter to see categorical columns too (count, unique, top, freq).

````python
df.describe(include='all')
````

## Step 4 — Find the missing values — `df.isnull().sum()`

`df.isnull()` returns a DataFrame of True/False — True wherever a value is missing. `.sum()` counts the Trues per column.

In plain English you might ask the agent: 
>“For each column in the DataFrame, count how many cells are empty or missing.”

**Why this matters before any analysis:** if you want to group things, you might run a `groupby()` on a column that has nulls. Pandas silently drops those rows. You may not notice — but your results will be wrong.

````python
df.isnull().sum()
````

### Track rows with missing `device_type`

Every row where `device_type` is missing is copied into **`null_device_type`**. That keeps an audit trail: you can inspect these reviews, export them, or compare ratings to known-device rows — without changing the main `df` until you decide how to handle nulls.

````python
# Rows where device_type is missing (NaN or empty / whitespace-only string)
_dt = df["device_type"]
_missing = _dt.isna() | (_dt.astype(str).str.strip() == "")
null_device_type = df.loc[_missing].copy()

print(f"Rows with missing device_type: {len(null_device_type)}")
null_device_type
````

### So what? 
Good question. Finding missing values is only the first step. The “so what?” is: missing data can change your results, especially when you group or summarize.

If `device_type` has 63 missing values (which we now know it does!), you have a few reasonable choices:

1. **Keep them as their own category**
   
     Plain English for the agent: 
     >“Treat missing device types as `Unknown` so they still show up in the analysis.” This is often the best first move because it makes the missingness visible.

```python
     df['device_type'] = df['device_type'].fillna('Unknown')
     df.groupby('device_type')['rating'].mean()
```

2. **Drop rows with missing values**
   
     Plain English for the agent: 
     >“Only analyze rows where `device_type` is known.” This is okay if the missing rows are small in number and not important to the question. But here, 63 out of 500 is about 13%, so dropping them could affect the result.

```python
     df_clean = df.dropna(subset=['device_type'])
     df_clean.groupby('device_type')['rating'].mean()
```

3. **Ask whether the missingness matters**
   
     Before deciding, compare the missing rows to the non-missing rows. Plain English:
     > “Do reviews with missing device type have different ratings than reviews with known device type?”

```python
     df[df['device_type'].isnull()]['rating'].mean()
     df[df['device_type'].notnull()]['rating'].mean()
```
If they look different, dropping them may bias your analysis.

### A good rule of thumb:

_If the missing column is important to your question, don’t silently drop it. Either keep missing values as `Unknown`, or explain why you removed them._

## Step 5 — Distribution of a column — `value_counts()`

`value_counts()` counts how many times each unique value appears. It's the most useful single operation for categorical data.

Use it to answer: *What are the most common values? Is the data balanced or skewed toward one category?*

````python
# How many reviews per app?
df['app'].value_counts()
````

````python
df['device_type'].value_counts(dropna=False)
````

### What's the distribution of ratings?

````python
df['rating'].value_counts().sort_index()
````

---
## Step 6 — Filter to a subset — boolean indexing

`df[condition]` returns only the rows where the condition is True. The condition is written as a comparison inside the brackets.

This is the same as Excel's filter — but repeatable, stackable, and documented in your code.

Ask the agent:
>“Using pandas, show me only the reviews where rating is less than 2. Use df[condition] syntax and explain what the filtered DataFrame contains.”

````python
# Reviews with a rating of 2 or below
low_rated = df[df['rating'] <= 2]

print(f"{len(low_rated):,} low-rated reviews ({len(low_rated)/len(df):.1%} of total)")
low_rated.head()
````

You can stack two conditions with & (and) or | (or). Parentheses are required around each condition.

````python
df[(df['rating'] <= 2) & (df['app'] == 'Fieldkit')].head(10)
````

---
## Step 7 — Compare across groups — `groupby()`

`groupby()` splits the DataFrame by a categorical column, then applies an aggregation function (mean, count, sum, etc.) to a numeric column within each group.

This answers the sprint question: *How does [metric] vary by [category]?*

````python
# Average rating per app — sorted lowest to highest
df.groupby('app')['rating'].mean().round(2).sort_values()
````

````python
# More than one stat at once with .agg()
df.groupby('app')['rating'].agg(['mean', 'count', 'std']).round(2)
````

````python
# Groupby across two columns
df.groupby(['app', 'device_type'])['rating'].mean().round(2)
````

## The first-look checklist — summary

```python
df = pd.read_csv('your_file.csv')   # load

df.head()                            # first five rows
df.tail()                            # last five rows
df.info()                            # columns, types, non-null counts
df.describe()                        # numeric summary stats
df.describe(include='all')           # includes categoricals

df.isnull().sum()                    # missing value count per column
df.isnull().mean().mul(100).round(1) # missing as percentage

df['col'].value_counts()             # distribution of a categorical column
df['col'].value_counts(dropna=False) # include NaN in the count

df[df['col'] > value]                # filter rows
df[(df['col1'] > v1) & (df['col2'] == v2)]  # stacked filter

df.groupby('cat_col')['num_col'].mean()      # group and aggregate
df.groupby('cat_col')['num_col'].agg(['mean','count','std'])
```

Run these in order. Every time. On every new dataset.
