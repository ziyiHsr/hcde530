### Question 2 — What is the distribution of your most important column?

This block answers **Question 2**: **`department`** is the main way these records are organized in the museum collection, so we show how many artworks fall in each department (`value_counts`).

```python
df["department"].value_counts()
```

### Question 3 — Filter to a meaningful subset. What is in it?

This block answers **Question 3**: we keep only rows from **Asian Art**—a focused slice for browsing catalog entries from that department—and show its size and a preview.

```python
asian_art = df[df["department"] == "Asian Art"].copy()
print("Rows in subset:", len(asian_art))
asian_art.head(10)
```

### Question 4 — Group by a category and take the average of a numeric column.

This block answers **Question 4**: we **group by `department`** and compute the **mean of `objectID`** (the Met’s numeric object identifier). The average ID is mainly useful here as practice for `groupby` + `.mean()` on a numeric field; interpret with care.

```python
df.groupby("department")["objectID"].mean().round(1).sort_values(ascending=False)
```
